```python
#importing the needed libraries
import pandas as pd
import numpy as np
```

```python
#loading the data
data=('https://archive.ics.uci.edu/ml/machine-learning-databases/00374/energydata_complete.csv')
dt = pd.read_csv(data)

#preview the data
dt
```

```python
#as instructed dropping columns light and date 

dt.drop(columns=(['date', 'lights']), axis=1, inplace=True)

dt
```

### QUESTION 12

```python
x = dt['T2']
y = dt['T6']
```

```python
#reshaping the x data
array = np.array([x])
x_shaped = array.reshape(-1, 1)
```

```python
#importing train_test_split from scikit.learn
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_shaped, y, test_size=0.3)
```

```python
#importing LinearRegression() from scikit.learn
from sklearn.linear_model import LinearRegression

#assigning it to a variable
linear_model = LinearRegression()

#fit the model to the training dataset
linear_model.fit(x_train, y_train)

#obtain predictions
predicted_values = linear_model.predict(x_test)
```

```python
#R-Squared for linear model
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, predicted_values)
round(r2_score, 2)
```

### QUESTION 13

```python
#There is need to normalise our dataset to a common scale using the min max scaler first
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scale_fit = scaler.fit_transform(dt)
normalised_data = pd.DataFrame(scale_fit, columns=dt.columns)
features_df = normalised_data.drop('Appliances', axis=1)
appliance_target = normalised_data['Appliances']
```

```python
#Then splitting the dataset into the training and testing dataset. Recalling that the features and target variables had earlier been segmented the 
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features_df, appliance_target, test_size=0.3, random_state=42)
```

```python
#import LinearRegression() from scikit.learn
from sklearn.linear_model import LinearRegression

#assigning LinearRegression() to a variable
lin_model = LinearRegression()

#fit the model to the training dataset
lin_model.fit(x_train, y_train)

#obtain predictions
y_hat = lin_model.predict(x_test)
```

```python
#importing MAE for the linearregressor
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_hat)
round(mae, 2)
```

### QUESTION 14

```python
#Calculating Residual Sum of Squares for linear regression
rss = np.sum(np.square(y_test - y_hat))
round(rss, 2)
```

### QUESTION 15

```python
#feature weights
def get_weights_df(model, feat, col_name):
  #this function returns the weight of every feature
  weights = pd.Series(model.coef_, feat.columns).sort_values()
  weights_df = pd.DataFrame(weights).reset_index()
  weights_df.columns = ['Features', col_name]
  weights_df[col_name].round(3)
  return weights_df

linear_model_weights = get_weights_df(lin_model, x_train, 'Linear_Model_Weight')
linear_model_weights.sort_values(by='Linear_Model_Weight')
```
